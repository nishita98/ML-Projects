```python
# imports needed

import nltk
import gensim
import pandas as pd
nltk.download('punkt', quiet=True)
nltk.download('stopwords')
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer 
```

```python
#loading the csv dataset into panda dataframe
bts_dataset = pd.read_csv('bts.csv')

#finding total rows and columns through shape method
bts_dataset.shape
```

```python
#displaying top 10 records
bts_dataset.head(10)
```

```python
#checking for count of null fields
bts_dataset.isnull().sum()
```

```python
#dropping null and duplicate records and resetting the index
bts_dataset=bts_dataset.dropna().drop_duplicates()
bts_dataset = bts_dataset.dropna().reset_index(drop=True)
bts_dataset.isnull().sum()
```

```python
#convert strings to lowercase
bts_comments = bts_dataset['comment_text'].str.lower()
#storing the stopwords
stop_words = stopwords.words('english')
#initializing the lemmatizer function
lemmatizer = WordNetLemmatizer() 

#data processing function
def preprocess(text):
    data = word_tokenize(text)# Tokenize the string into words
    data = [word for word in data if word not in stop_words]# Filter out stopwords
    data = [word for word in data if word.isalpha()] # Remove non-alphabetic tokens, such as punctuation
    data = [lemmatizer.lemmatize(word) for word in data]#converting a word to its base form
    return data
```

```python
#calling the preprocess function on bts_comments column
processed_comments = [preprocess(text) for text in bts_comments]
```

```python
#displaying the processced comments
processed_comments
```

```python
#Load Word2Vec embedding model
#setting min_count parameter: Minimum word count threshold

model = gensim.models.Word2Vec(min_count=20)
```

```python
#building model vocabulary
model.build_vocab(processed_comments, progress_per=1000)
```

```python
#training the model
#sentences parameter: processed_comments
#total_examples: count of sentences of model
#epochs: number of iterations on model
model.train(processed_comments, total_examples=model.corpus_count, epochs=model.epochs)
```

```python
#saving the model
model.wv.save_word2vec_format('bts.bin')
```

```python
#finding most similar vector representations
model.wv.most_similar("jin") #jin is BTS hyung i.e. oldest member
model.wv.most_similar('ugh') #ugh and others listed in the output are BTS songs mentioned in the comments
```

```python
#finding analogy difference
model.wv.most_similar(positive=["kim", "min"], negative=["tae"]) #Kim Tae, #Min Yoongi
```

```python
#testing similarity between two words
model.wv.similarity('bts', 'army') #output - 0.94627094:  Army are BTS fans
model.wv.similarity('min', 'yoongi') #output - 0.97752297: Yoongi's surname is Min
model.wv.similarity('min', 'jungkook') #output - 0.5365919: Jungkook's surname is Jeon
```

```python
#finding out odd one out
model.wv.doesnt_match(['txt', 'bts', 'army']) #output - txt: Army are BTS fans
model.wv.doesnt_match(['seokjin', 'jungkook', 'v'])#output - seokjin: Jungkook and V are best friends and youngest
```
